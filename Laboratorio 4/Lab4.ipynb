**Respuestas**

1.	Apache Spark es un motor de procesamiento distribuido dise침ado para manejar grandes vol칰menes de datos. Sus principales ventajas sobre Pandas son la capacidad de procesar datos en paralelo y en cl칰steres distribuidos, permitiendo manejar datasets mucho m치s grandes y complejos, mientras que Pandas se limita a datos que pueden caber en memoria.Tambi칠n es m치s r치pido que Pandas porque puede distribuir el trabajo en m칰ltiples computadoras.

2.	Un RDD (Resilient Distributed Dataset) es una colecci칩n distribuida e inmutable de elementos que constituye el pilar fundamental de Apache Spark, este modelo permite que Spark realice operaciones de computaci칩n en paralelo de manera eficiente. Su principal caracter칤stica son las dependencias  y la partici칩n autom치tica en cl칰steres. Los dataframes en Spark son una abstracci칩n m치s moderna que se construye sobre los RDD, proporcionando una API m치s amigable y optimizaciones adicionales.

3.  Las transformaciones son operaciones que generan un nuevo DataFrame a partir de uno existente sin modificar el conjunto de datos original, un ejemplo de transformaci칩n es la funci칩n filter() la cual se encarga de filtrar los datos dada una condici칩n. Por otro lado, las acciones son las operaciones que desencadenan la ejecuci칩n de todas las transformaciones acumuladas y devuelven los resultados obtenidos, un ejemplo ser칤a la funci칩n count(), la cual se utiliza para calcular el n칰mero total de registros en un RDD o DataFrame. Cuando se ejecuta una acci칩n en Spark, se construye un DAG que optimiza y divide el plan en etapas. Luego, distribuye las tareas a los ejecutores para procesar los datos en paralelo, y los resultados se env칤an al driver, completando la ejecuci칩n.

4.	El particionamiento en Spark es clave para dividir los datos entre nodos en un cl칰ster, afectando directamente el rendimiento. Un buen particionamiento balancea la carga de trabajo entre nodos, reduciendo la sobrecarga de comunicaci칩n y maximizando el procesamiento paralelo.

5.	El Spark Driver coordina la ejecuci칩n de las aplicaciones, traduciendo el c칩digo en tareas distribuidas. Los Spark Executors son los encargados de ejecutar las tareas en los nodos del cl칰ster, almacenando y procesando los datos asignados por el Driver.

6.	El Catalyst Optimizer es el optimizador de consultas SQL en Spark. Su funci칩n principal es mejorar el rendimiento de las consultas transformando y reorganizando los planes de ejecuci칩n, utilizando reglas de optimizaci칩n y estad칤sticas para generar planes eficientes.

Observamos que todas las variables del tipo `string` est치n bien clasificadas. "Transaction Date" tambi칠n ya que es del tipo `timestamp_ntz` que es hora en Pyspark. Por lo tanto, podemos hacer correcciones en los tipos `double`, ya que "Quantity", "Customer Age", "Is An Alien", "Account Age Days" y "Transaction Hour" pueden ser enteros (`int`). Los datos que pueden ser del tipo `double` (decimal) es "Transaction Amount" (obligatoriamente) y "Transaction Hours".

````python
# 5. Crear funci칩n cast_columns que cambie el tipo de una o m치s columnas
def cast_columns(df, cols_types):
    """
    Cambia el tipo de m칰ltiples columnas en un DataFrame de Spark.

    Par치metros:
    - df: DataFrame de Spark.
    - cols_types: Diccionario con nombres de columnas como claves y tipos de datos de Spark como valores.

    Retorna:
    - DataFrame de Spark con tipos de columnas modificados.
    """
    # referencia de la clase
    for col_name, new_type in cols_types.items():
        df = df.withColumn(col_name, df[col_name].cast(new_type))
    return df

# cambiamos los tipos de las columnas necesarias
cols_to_cast = {
    "Quantity": IntegerType(),          # cantidad: num. entero
    "Customer Age": IntegerType(),      # edad: num. entero
    "Is An Alien": IntegerType(),       # booleando: num. entero (0 o 1)
    "Account Age Days": IntegerType()  # d칤as: num. entero
    }

df = cast_columns(df, cols_to_cast)
df.printSchema()     # observamos los cambios
````

````python
# 6. Contar nulos por variables

from pyspark.sql.functions import when, rand, col, sum

# se recorre cada columna (c), 
# luego, se crea una expresi칩n condicional con 'when' que es 1 si en la fila hay nulo y 0 si no
# suma todos los 1's (nulos) de la columna y con 'alias' le asigna el nombre de la columa
exprs = [sum(when(col(c).isNull(), 1).otherwise(0)).alias(c) for c in df.columns]
df.agg(*exprs).show()    # aplica lista 'exprs' con el m칠todo agregaci칩n ('agg') y muestra ('show')
````

````python
# 7. Eliminar datos nulos
df = df.na.drop()
````

1.   Agregar una columna llamada "Transaction bp" con el **monto total** de la transacci칩n en bodoque pesos. Se considera que $x$ d칩lares equivalen a $log(48+|x^{36}|)$ bodoque pesos.
2.   Crear una columna llamada "Transaction Month" con el mes en que se realiza una transacci칩n.
3.   Crear la variable *Type of purchase* seg칰n la catidad de unidades vendidas de acuerdo a las siguientes categor칤as.
  * Compra minorista: 4 productos o menos.
  * Compra mayorista: 5 produtos o m치s.
4. Imprima los registros de compras hechas por alien칤genas en el comecio mayorista.  Utilice `.filter()`.
5. Cuente la cantidad de compras realizadas por humanos y la cantidad de compras realizadas por alien칤genas. Utilice `.groupby()`.
6. Muestre una tabla con la recaudaci칩n promedio por transacci칩n para cada m칠todo de pago, tanto para humanos como alien칤genas. Utilice `pivot()`.

````python
# 1. Agregar columna "Transaction bp"
from pyspark.sql.functions import col, log, abs, pow, round  # funciones a usar

# Aplicar la f칩rmula y crear la nueva columna "Transaction bp"
df = df.withColumn("Transaction bp", log(48 + abs(pow(col("Transaction Amount"), 36))))
df = df.withColumn("Transaction bp", round(col("Transaction bp"),2))    # redondear

df.select("Transaction Amount", "Transaction bp").show(5)  # mostrar primeras filas
````

````python
# 2. Crear columna "Transaction Month"
from pyspark.sql.functions import month

df = df.withColumn("Transaction Month", month(col("Transaction Date")))
df.select("Transaction Date", "Transaction Month").show(5)   # mostrar primeras filas
````

````python
# 3. Crear columna "Type of purchase" seg칰n unidades vendidas
df = df.withColumn("Type of purchase", 
                   when(col("Quantity") <= 4, "Compra minorista")  # condici칩n, si se cumple, se completa
                   .otherwise("Compra mayorista"))                 # resultado en caso contrario

df.select("Quantity", "Type of purchase").show(5)
````

````python
# 4. Registros de compras hechas por alien칤genas en el comecio mayorista
df_aliens_mayorista = df.filter((col("Is An Alien") == 1) & (col("Type of purchase") == "Compra mayorista"))
df_aliens_mayorista.show(5)
````

````python
# 5. Contar la cantidad de compras realizadas por humanos y las de alien칤genas
from pyspark.sql.functions import count

# agrupar por "Is An Alien" y contar la cantidad de transacciones, nombra la col como "Total Compras"
df_grouped = df.groupBy("Is An Alien").agg(count("*").alias("Total Compras"))

# agregar aclaraci칩n de quien compra
df_grouped = df_grouped.withColumn(
    "Tipo de Compra", 
    when(col("Is An Alien") == 1, "Compra alien")
    .otherwise("Compra humano")
)

df_grouped.show()
````

````python
# 6. Tabla con la recaudaci칩n promedio por transacci칩n para cada m칠todo de pago

# Agrupar por m칠todo de pago y realizar un pivot para alien칤genas y humanos
df_pivot = df.groupBy("Payment Method").pivot("Is An Alien").mean("Transaction Amount")

# Renombrar las columnas resultantes del pivot (0: humanos, 1: alien칤genas)
df_pivot = df_pivot.withColumnRenamed("0", "Prom Humanos (0)").withColumnRenamed("1", "Prom Alien칤genas (1)")

# Mostrar el resultado
df_pivot.show()
````

````python
# Extra: Agregamos una columna llamada "Comprador"

# Crear la columna "Comprador" que ser치 "Alien" si "Is An Alien" == 1, y "Humano" si es 0
df = df.withColumn(
    "Comprador", when(df["Is An Alien"] == 1, "Alien").otherwise("Humano")
)

# Ver primeros registros para verificar
df_aliens = df.select("Is An Alien", "Comprador")
df_aliens.sample(fraction=0.05, seed=2).show(7)
````

Observamos que el m칠todo del objeto es `LogisticRegression` de `scikit-learn` y tiene 2 m칠todos de predicciones:

1. `predict(self, X)`: devuelve las clases predichas (return `y_pred`: Vector containing the class labels for each sample)
2. `predict_proba`: devuelve las probabilidades estimadas de las clases predichas. (return `T`: Returns the probability of the sample for each class in the model, where classes are ordered as they are in `self.classes_`)

````python
# 3. Crear una funci칩n personalizada que prediga la probabilidad de fraude utilizando el 
#    칰ltimo valor de la lista generada por el objeto serializado

def predict_fraude(df_columnas):
    # Obtener el 칰ltimo valor de las caracter칤sticas
    ultimo_valor = df_columnas.iloc[[-1]]
    
    # Predecir la probabilidad de fraude utilizando el modelo
    probabilidad_fraude = obj_serializado.predict_proba(ultimo_valor)[:, 1]  # Probabilidad de clase 1 (fraude)
    
    return probabilidad_fraude
````
